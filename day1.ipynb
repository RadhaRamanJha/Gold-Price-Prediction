```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import warnings
warnings.filterwarnings('ignore')
from statsmodels.tsa.seasonal import STL
```

```python
gold_d=pd.read_csv("Gold_Price_in_India_for_Project.csv",parse_dates=['Date'])
```

```python
gold_d
```

```python
gold_d.info
```

```python
gold_d =gold_d.set_index(pd.DatetimeIndex(gold_d['Date'])).drop('Date',axis=1)
```

```python
gold_d.isna().sum()
```

```python
gold_d.dtypes
```

```python
gold_d["Date"]=pd.to_numeric(gold_d["Date"],errors="coerce")
```

```python
gold_d.dtypes
```

```python
gold_d = gold_d.rename(columns = {"Indian rupee (INR)":"Inr"})
```

```python
gold_d.shape
```

```python
sns.boxplot(gold_d["Inr"])
```

```python
gold_d.plot(figsize = (15,5))
plt.xlabel("Observation Date")
plt.ylabel("Prices")
plt.title("Gold Prices over Time")
plt.show()
```

```python
gold_d.plot(xlim = ['2021-08-21','2021-12-21'],ls='--', c='g',ylim = [100000,150000],figsize = (15,5))
plt.xlabel("Observation Date")
plt.ylabel("Prices")
plt.title("Gold Prices in Last Four months")
plt.show()
```

```python
last_4month_index = gold_d.loc['2021-08-21':'2021-12-21'].index
last_4month_price = gold_d.loc['2021-08-21':'2021-12-21']['Inr']
```

```python
figure,ax = plt.subplots(figsize = (15,5))
plt.tight_layout()
figure.autofmt_xdate()
ax.plot(last_4month_index,last_4month_price)
plt.show()
```

```python
_,ax = plt.subplots(figsize=(25,25))
sns.boxplot(x=gold_d.index.year, y = gold_d.Inr,ax=ax)
plt.title("Box Plot of Gold Prices from 2016 to 2021")
plt.xlabel("Year")
plt.ylabel("Gold Prices")
plt.show()
```
